# Bukelwa Moerane's Crime Stats Project

Here I am giving insight from the HOUSEHOLD data. 

```python
import pandas as pd
import matplotlib.pyplot as plt 

file = pd.read_csv('VOCS2016-2017-HOUSEHOLD_F1.csv')
```

# How many people were surveyed

```python
# Here we are trying to show how many people where sureveyed in each province

import pylab as pyl
locations = file.groupby('prov') .size() .reset_index(name = 'Number of people surveyed')
locations

labels = ['Western Cape ', 'Eastern Cape', 'Northern Cape ','Free State', 'KwaZulu-Natal','North West ', 'Gauteng', 'Mpumalanga','Limpopo']
xaxis = locations['prov']
yaxis = locations['Number of people surveyed']

bar = plt.bar(xaxis, yaxis)
plt.ylabel('Cases/ Number of People Interviewed')
plt.xlabel('Provinces ')
pyl.xticks(locations['prov'], labels, rotation = 80)
bar[6].set_color('red')
fig = plt.gcf()
fig.set_size_inches(10,10)
pyl.show()
```

In my observation there are a lot of people who were interviewed in Gauteng. This enitre survey could be a bit bias, because they interviewed a lot of people from gauteng compared to other provinces. How will we know which province experiences the most crime.

# Exracting Gauteng

In this section I am showing all rows that belong to the Gauteng province.

```python
gauteng = file[file.prov == 7]
```

# How long has the interviewee been staying in Gauteng

```python
staying = gauteng.groupby('Q21bHowLong') .size() .reset_index(name = 'Number of people surveyed')
months = staying[0:3]
months

labels = ['Less than 12 months', '12  23 months ', '24 months or more']
width = 1/2.5

bargraph = plt.bar(months['Q21bHowLong'], months['Number of people surveyed'], width)
plt.ylabel('Number of People')
plt.xlabel('Number of Months ')
plt.title('Number of people who live in Gauteng for period of time')
pyl.xticks(months['Q21bHowLong'], labels, rotation = 50)
fig = plt.gcf()
fig.set_size_inches(10,10)
pyl.show()
    
    
```

# Crime Topic

How many people in Gauteng talk about crime in their household 

```python
CrimeTopic = gauteng.groupby('Q21topiccrime') .size()
CrimeTopic
CrimeTopic = CrimeTopic[0:2]

labels = ['Yes', 'No', ]
color = ['palegreen','crimson']

#show percentage for each group on the pie chart 

plt.pie(CrimeTopic, labels =labels, colors = color, autopct='%1.0f%%')
plt.title('How Many People Talk About Crime')
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.show()
```

# Change in Life style

Here we are trying to figure out how many people have changed their behaviour and routes because they are afraid of crime. The app could help them feel more secure and allow them the freedom to behave and travell freely. 

```python
changeofbehaviour = gauteng[['Q1119Changed', 'Q1717Changebehaviour']]
avoidroute = gauteng[['Q11204Avoid']]
yes = changeofbehaviour[changeofbehaviour == 1] .count() .sum()
no  = changeofbehaviour[changeofbehaviour == 2] .count() .sum()
yes1 = avoidroute[avoidroute == 1] .count() .sum()
no1  = avoidroute[avoidroute == 2] .count() .sum()

yes = yes + yes1
no = no + no1
plot = yes, no
labels = ['Yes', 'No']
color = ['palegreen','crimson']

plt.pie(plot, labels =labels, autopct='%1.0f%%', colors = color)
plt.title('How Many Poeple Changed Routes and Behaviour')
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.show()
```

# The Emergency Call
