# Bukelwa Moerane's Crime Stats Project

In this project we were required to import a database with statistics on crime in South Africa. We live in Johannesburg, so what we need to do is extract and get insight from this database on crime in Gauteng. 

<i syle = "font-size:10px;">Stats South AFrica conducted a survey 2016 - 2017 to figure out the most frequent crimes that occur in South Africa.</i> 

```python
#here we import our DataFrame using Pandas

import pandas as pd
import matplotlib.pyplot as plt 

file = pd.read_csv('VOCS2016-2017-SECTION21TO28_F1.csv')
file
```

# How many people were surveyed

```python
# Here we are trying to show how many people where sureveyed in each province

import pylab as pyl

locations = file.groupby('prov') .size() .reset_index(name = 'Number of people surveyed')
locations

labels = ['Western Cape ', 'Eastern Cape', 'Northern Cape ','Free State', 'KwaZulu-Natal','North West ', 'Gauteng', 'Mpumalanga','Limpopo']
bar = plt.bar(locations['prov'],locations['Number of people surveyed'])
plt.ylabel('Cases/ Number of People Interviewed')
plt.xlabel('Provinces ')
pyl.xticks(locations['prov'], labels, rotation = 50)
bar[6].set_color('red')
pyl.show()
```

In my observation there are a lot of people who were interviewed in Gauteng. This enitre survey could be a bit bias, because they interviewed a lot of people from gauteng compared to other provinces. How will we know which province experiences  the most crime. 

# Exracting Gauteng

In this section I am showing all rows that belong to the Gauteng province.

```python
gauteng = file[file.prov == 7]
gauteng
```

# Insight From Gauteng Survey

# Gender 

Taking a closer look at how many females and males were interviewed in Gauteng.

```python
gender = gauteng.groupby('C_GENDER') .size()
gender

labels = ['males', 'Females']
color = ['lightpink','deepskyblue' ]

#show percentage for each group on the pie chart 

plt.pie(gender, labels =labels, colors = color, autopct='%1.0f%%' )
plt.title('Different Gender Interviewed')
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.show()
```

# Looking at age groups

Taking a closer look at different age groups in Gauteng.

```python
age = gauteng.groupby('D_AGE') .size() .reset_index(name = 'Number of people interviewed')
age

plt.plot(age['D_AGE'], age['Number of people interviewed'])
plt.ylabel('Number of people interviewed')
plt.xlabel('AGE')
plt.title('Ages of people interviewed: Gauteng')
plt.rcParams['figure.figsize'] = (20,16)
plt.show()
```

# Let us take a look at the different population groups

Again I am afriad that the number of blacks interviewed could lead us to making bias conclusions. We need to have a fair or equal number of people interviewed per population group, so that we can make a valid conclusion on which population group experiences the most crime. 

The big question that this kind of data will answer is: " Which population group experiences the most crime?" "Why that specific population or racial group?"

```python
population = gauteng.groupby('E_POPULATION') .size() 
population

labels = ['Black African', 'Coloured', 'Indian/Asian', 'White ']


plt.pie(population, labels =labels, autopct='%1.0f%%' )
plt.title('Different Racial/ Population Groups Interviewed')
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.show()
```

# Overview of Crimes


```python
crimes  = gauteng[['Q21b1aPersProp', 'Q21b11CarHijack', 'Q21b11cRob', 'Q21b1dSexual', 'Q21b1eAssault', 'Q21b1hOther']]
crimes
```

```python
# Here I am grouping each crime to figure out how many people reported yes to experiencing each crime

persprop = crimes.groupby('Q21b1aPersProp') . size()
CarHij = crimes.groupby('Q21b11CarHijack') .size()
Rob = crimes.groupby('Q21b11cRob') .size()
SexualOff = crimes.groupby('Q21b1dSexual') .size()
Assualt = crimes.groupby('Q21b1eAssault') .size()
Other = crimes.groupby('Q21b1hOther') .size()

print(persprop,CarHij,SexualOff,Assualt, Other )
```

```python
#Here I am going to Plot the above grouped information in a bar graph

#These are all the "YES" cases
x_axis = ['Personal property','Car Hijacking','Sexual Offense','Assualt', 'Other']
y_axis = [persprop[1],CarHij[1],SexualOff[1],Assualt[1], Other[1]]

plt.bar(x_axis, y_axis)
plt.ylabel('Occurance')
plt.xlabel('Types Of Crimes')
plt.title("Crimes That Were Experienced By People In Gauteng")
plt.show()
```

These are the people who reported 'NO' to experiencing the crime

```python
#These are all the "NO" cases

x_axis = ['Personal property','Car Hijacking','Sexual Offense','Assualt', 'Other']
y_axis = [persprop[2],CarHij[2],SexualOff[2],Assualt[2], Other[2]]

plt.bar(x_axis, y_axis)
plt.ylabel('Occurance')
plt.xlabel('Types Of Crimes')
pyl.xticks(x_axis, rotation = 40)
plt.title("Crimes That Were Experienced By People In Gauteng")
plt.show()
```

Now to compare "YES" cases and "NO" cases.

```python

x_axis = ['Personal property','Car Hijacking','Sexual Offense','Assualt', 'Other']
y_axis = [persprop[2],CarHij[2],SexualOff[2],Assualt[2], Other[2]]
y1_axis = [persprop[1],CarHij[1],SexualOff[1],Assualt[1], Other[1]]
width = 1/1.5

plt.bar(x_axis, y_axis, width)
plt.bar(x_axis, y1_axis, width)
plt.ylabel('Occurance')
plt.xlabel('Types Of Crimes')
pyl.xticks(x_axis, rotation = 40)
plt.title("Crimes That Were Experienced By People In Gauteng")
plt.show()
```
